```python
import numpy as np
import sqlite3
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
import glob
```

```python
files = glob.glob('*10yrs.db')
files.sort()

for filename in files:

    con = sqlite3.connect(filename)
    df = pd.read_sql("select night from observations where filter='u' and note not like '%DD%'", con)
    con.close()
    yval = np.arange(df['night'].size)
    plt.plot(df['night'], yval, label=filename.replace('10yrs.db', ''))
plt.legend()
```
